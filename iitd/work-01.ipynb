### Check linear dependence / independence of |+> and |->
#### Method 1
If two vectors are linearly dependent then the following must be true.

$m|+> + n|-> = 0$ ...  (1)

$|+> = \begin{bmatrix}1/\sqrt2 \\ 1/\sqrt2 \end{bmatrix}$ &nbsp; $|-> = \begin{bmatrix}1/\sqrt2 \\ -1/\sqrt2 \end{bmatrix}$

Replace the values of basis vectors $|+>$ and $|->$ in (1)

$m \begin{bmatrix}1/\sqrt2 \\ 1/\sqrt2 \end{bmatrix} + n \begin{bmatrix}1/\sqrt2 \\ -1/\sqrt2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $

=> $ \begin{bmatrix} (m+n)/\sqrt2 \\ (m-n)/\sqrt2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $

That gives us two equations.

$ m = -n$  
$m = n $

Since the equations above are inconsistent, it gives us that $|+>$ and $|->$ are linearly independent

#### Method 2
If two vectors are linearly independent then the following must be true.

$m|+> + n|-> = |v>$ ...  (2)

$|+> = \begin{bmatrix}1/\sqrt2 \\ 1/\sqrt2 \end{bmatrix}$ &nbsp; $|-> = \begin{bmatrix}1/\sqrt2 \\ -1/\sqrt2 \end{bmatrix}$

Replace the values of basis vectors $|+>$ and $|->$ in (2)

$m \begin{bmatrix}1/\sqrt2 \\ 1/\sqrt2 \end{bmatrix} + n \begin{bmatrix}1/\sqrt2 \\ -1/\sqrt2 \end{bmatrix} = \begin{bmatrix} a \\ b \end{bmatrix} $

Convert the above to linear equations.

$ 
m (1/\sqrt2) + n (1/\sqrt2) = a \\ 
m (1/\sqrt2) + n (-1/\sqrt2) = b
$

Translate that to augemented matrix.

$
  \left[\begin{array}{rr|r}
    1/\sqrt2 & 1/\sqrt2 & a \\
    1/\sqrt2 & -1/\sqrt2 & b \\
  \end{array}\right]
$

Perform $ R1 = R1 * \sqrt2 $

$
  \left[\begin{array}{rr|r}
    1 & 1 & a\sqrt2 \\
    1/\sqrt2 & -1/\sqrt2 & b \\
  \end{array}\right]
$

Perform $ R2 = R2 - (1/\sqrt2)R1 $

$
  \left[\begin{array}{rr|r}
    1 & 1 & a\sqrt2 \\
    0 & -\sqrt2 & b-a \\
  \end{array}\right]
$

Perform $ R2 = R2 * (-1/\sqrt2) $

$
  \left[\begin{array}{rr|r}
    1 & 1 & a\sqrt2 \\
    0 & 1 & (a-b)/\sqrt2 \\
  \end{array}\right]
$

Perform $ R1 = R1 - R2 $

$
  \left[\begin{array}{rr|r}
    1 & 0 & (a+b)/\sqrt2 \\
    0 & 1 & (a-b)/\sqrt2 \\
  \end{array}\right]
$

We can take any a,b and express them back into coefficients m and n. 
Any point in the span can be respresented with |+> and |->. 
Hence they are linearly independent.
