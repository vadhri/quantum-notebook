```python
import pennylane as qml
import numpy as np
```

```python
dev = qml.device('default.qubit', wires=2)

@qml.qnode(dev)
def prepare_psi_plus():
    qml.Hadamard(0)
    qml.CNOT(wires=[0,1])
    return qml.state()


@qml.qnode(dev)
def prepare_psi_minus():
    qml.PauliX(0)
    qml.Hadamard(0)
    qml.CNOT(wires=[0,1])
    
    return qml.state()


@qml.qnode(dev)
def prepare_phi_plus():
    qml.PauliX(1)
    qml.Hadamard(0)
    qml.CNOT(wires=[0,1])    
 
    return qml.state()


@qml.qnode(dev)
def prepare_phi_minus():
    qml.PauliX(0)
    qml.PauliX(1)
    qml.Hadamard(0)
    qml.CNOT(wires=[0,1])    
    
    return qml.state()


psi_plus = prepare_psi_plus()
psi_minus = prepare_psi_minus()
phi_plus = prepare_phi_plus()
phi_minus = prepare_phi_minus()

# Uncomment to print results
print(f"|ψ_+> = {psi_plus}")
print(f"|ψ_-> = {psi_minus}")
print(f"|ϕ_+> = {phi_plus}")
print(f"|ϕ_-> = {phi_minus}")
```

```python
qml.draw_mpl(prepare_psi_minus)()
```

```python
qml.draw_mpl(prepare_phi_plus)()
```

```python
qml.draw_mpl(prepare_phi_minus)()
```
