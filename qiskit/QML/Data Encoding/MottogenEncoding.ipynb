### Amplitude encoding using mottogen state preparation method using qiskit.

```python
from qiskit import *
import pennylane as qml
import numpy as np
from qiskit.quantum_info import *
```

```python
dev = qml.device('default.qubit', wires=3)

@qml.qnode(dev)
def circuit(state):
    qml.MottonenStatePreparation(state_vector=state, wires=range(3))
    return qml.state()

state = np.array([0.2, 0, 0.5, 0, 0, 0, 0.2, 0.1])
state = [math.sqrt(i) for i in state]
print (state)
print(qml.draw_mpl(circuit, decimals=5, expansion_strategy="device", max_length=80)(state))
```

```python
QC = QuantumCircuit(3)
QC.ry(1.15928, 0)
QC.ry(2.57765, 1)
QC.ry(0.30774, 2)
QC.barrier()
QC.cx(0,1)

QC.ry(-0.56394, 1)
QC.cnot(0,1)
QC.cnot(1,2)
QC.ry(-0.30774, 2)
QC.barrier()
QC.cnot(0,2)
QC.ry(0.30774, 2)
QC.cnot(1,2)
QC.ry(-0.30774, 2)
QC.cnot(0,2)

QC.draw(output="mpl", scale=0.75)
```
