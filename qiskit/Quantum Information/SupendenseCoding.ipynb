# Two qubits are prepared in bell state. Alice gets the first Qubit state, bob gets second. 

When alice receives the state she encodes the message into the circuit and bob would receive it at the end circuit resultant state. 

|Message|Gate Applied|
|-|-|
|00|I|
|01|X|
|10|Z|
|11|ZX|

## 01 

```python
from qiskit import *
from qiskit.quantum_info import Statevector

Alice_message = QuantumCircuit(1, name="Alice-Message-Encoder")

Alice_message.x(0)
Alice_message.to_gate()

QC = QuantumCircuit(2,2)

QC.h(0)
QC.cx(0,1)
QC.barrier()

QC.append(Alice_message, [0])

QC.barrier()

QC.cx(0,1)
QC.h(0)

QC.draw(output="mpl")

```

```python
state = Statevector(QC)
state.draw("latex")
```

```python
from qiskit import *
from qiskit.quantum_info import Statevector

Alice_message = QuantumCircuit(1, name="Alice-Message-Encoder")

Alice_message.x(0)
Alice_message.z(0)

Alice_message.to_gate()

QC = QuantumCircuit(2,2)

QC.h(0)
QC.cx(0,1)
QC.barrier()

QC.append(Alice_message, [0])

QC.barrier()

QC.cx(0,1)
QC.h(0)

QC.draw(output="mpl")

```
