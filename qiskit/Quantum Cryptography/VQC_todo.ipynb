```python
from qiskit.circuit import *
import numpy as np

x = Parameter('X')
theta_list = ParameterVector('t', length=3)

qc = QuantumCircuit(1,1)
qc.rx(x,0)
qc.ry(theta_list[0], 0)
qc.rz(theta_list[1], 0)
qc.rx(theta_list[2], 0)

qc.measure(0,0)
qc.draw(output="mpl", scale=0.5)
```

```python
initial_point = np.array([0.4, 0.09, 0.42, 0.34])
encode_circuit = qc.bind_parameters(initial_point)
encode_circuit.decompose().decompose().draw(output="mpl", scale=0.5)
```
