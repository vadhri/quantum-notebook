# BB84 protocol.

https://en.wikipedia.org/wiki/BB84

```python
from qiskit import *
from qiskit.visualization import *
```

### Objective 1; Transfer 10 bits data from Alice to bob

```python
QuantumBB84Circuit = QuantumCircuit(10,10)
QuantumBB84Circuit.draw()
```

```python
# State to transfer.
QuantumBB84Circuit.x(0)
QuantumBB84Circuit.x(2)
QuantumBB84Circuit.x(4)
QuantumBB84Circuit.x(6)
QuantumBB84Circuit.x(8)
QuantumBB84Circuit.barrier()

QuantumBB84Circuit.draw(output="mpl")
```

### Alice random encoding. 


```python
QuantumBB84Circuit.h(0) 
QuantumBB84Circuit.z(1) 
QuantumBB84Circuit.z(2) 
QuantumBB84Circuit.h(3) 
QuantumBB84Circuit.z(4) 
QuantumBB84Circuit.h(5) 
QuantumBB84Circuit.z(6) 
QuantumBB84Circuit.z(7) 
QuantumBB84Circuit.h(8) 
QuantumBB84Circuit.z(9) 

QuantumBB84Circuit.barrier()
```

```python
QuantumBB84Circuit.draw(output="mpl")
```

### Bob;s random choice.

```python
QuantumBB84Circuit.h(0) 
QuantumBB84Circuit.z(1) 
QuantumBB84Circuit.h(2) 
QuantumBB84Circuit.z(3) 
QuantumBB84Circuit.h(4) 
QuantumBB84Circuit.z(5) 
QuantumBB84Circuit.h(6) 
QuantumBB84Circuit.h(7) 
QuantumBB84Circuit.z(8) 
QuantumBB84Circuit.h(9) 

QuantumBB84Circuit.barrier()
QuantumBB84Circuit.draw(output="mpl")
```
