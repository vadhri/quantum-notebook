# BB84 protocol.

https://en.wikipedia.org/wiki/BB84

```python
from qiskit import *
from qiskit.visualization import *
```

### Objective 1; Transfer 10 bits data from Alice to bob

```python
QuantumBB84Circuit = QuantumCircuit(10,10)
QuantumBB84Circuit.draw()
```

```python
# State to transfer.
QuantumBB84Circuit.x(0)
QuantumBB84Circuit.x(2)
QuantumBB84Circuit.x(4)
QuantumBB84Circuit.x(6)
QuantumBB84Circuit.x(8)
QuantumBB84Circuit.barrier()

QuantumBB84Circuit.draw(output="mpl")
```

### Alice random encoding. 


```python
QuantumBB84Circuit.h(0) # or qc_ab.z(0) # switch these based on your own choice
QuantumBB84Circuit.z(1) # or qc_ab.h(1)
QuantumBB84Circuit.z(2) # or qc_ab.h(2)
QuantumBB84Circuit.h(3) # or qc_ab.z(3)
QuantumBB84Circuit.z(4) # or qc_ab.h(4)
QuantumBB84Circuit.h(5) # or qc_ab.z(5)
QuantumBB84Circuit.z(6) # or qc_ab.z(5)
QuantumBB84Circuit.z(7) # or qc_ab.z(5)
QuantumBB84Circuit.h(8) # or qc_ab.z(5)
QuantumBB84Circuit.z(9) # or qc_ab.z(5)

QuantumBB84Circuit.barrier()
```

```python
QuantumBB84Circuit.draw(output="mpl")
```
